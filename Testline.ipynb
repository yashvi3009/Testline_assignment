```python
import json
import pandas as pd

# Load Current Quiz Submission Datasets
with open("/content/data.json", "r") as f:
    data = json.load(f)

with open("/content/quiz_submission.json", "r") as f:
    quiz_submission = json.load(f)

# Load Historical Quiz Data
with open("/content/historic.json", "r") as f:
    historic_data = json.load(f)

df_submission_1 = pd.DataFrame(data)
df_submission_2 = pd.DataFrame([quiz_submission])
df_historical = pd.DataFrame(historic_data)
```

**Descriptive Statistics**

```python
print(df_submission_2.describe())

# Unique quizzes and users
print("Unique quizzes:", df_submission_2['quiz_id'].nunique())
print("Unique users:", df_submission_2['user_id'].nunique())

# counts per quiz
submission_counts = df_submission_2['quiz_id'].value_counts()
print("Submissions per quiz:\n", submission_counts.head())
```

```python
# Calculate 'time_taken' if 'started_at' and 'ended_at' columns exist
if 'started_at' in df_submission_2.columns and 'ended_at' in df_submission_2.columns:
    try:
        df_submission_2['started_at'] = pd.to_datetime(df_submission_2['started_at'])
        df_submission_2['ended_at'] = pd.to_datetime(df_submission_2['ended_at'])
        df_submission_2['time_taken'] = (df_submission_2['ended_at'] - df_submission_2['started_at']).dt.total_seconds()
        time_taken_distribution = df_submission_2['time_taken'].describe()
        print("Time Taken Distribution:\n", time_taken_distribution)
    except ValueError as e:
        print(f"Error converting timestamps: {e}")
else:
    print("Columns 'started_at' or 'ended_at' not found in the DataFrame.")

# Analyze score distribution
if 'score' in df_submission_2.columns:
    score_distribution = df_submission_2['score'].describe()
    print("\nScore Distribution:\n", score_distribution)
else:
    print("\nNo 'score' column found in the DataFrame.")

# Analyze accuracy distribution
if 'accuracy' in df_submission_2.columns:
    accuracy_distribution = df_submission_2['accuracy'].describe()
    print("\nAccuracy Distribution:\n", accuracy_distribution)
else:
    print("\nNo 'accuracy' column found in the DataFrame.")

# Analyze quiz submissions
if 'quiz_id' in df_submission_2.columns:
    submission_counts = df_submission_2['quiz_id'].value_counts()
    print("\nQuizzes with the highest number of submissions:\n", submission_counts.head())
    print("\nQuizzes with the lowest number of submissions:\n", submission_counts.tail())
    quizzes_zero_submissions = submission_counts[submission_counts == 0]
    if not quizzes_zero_submissions.empty:
        print("\nQuizzes with zero submissions:\n", quizzes_zero_submissions)
else:
    print("\nNo 'quiz_id' column found in the DataFrame.")
```

```python
import pandas as pd
import numpy as np
import json

# Initialize insights dictionary
insights = {}

# 1. Analyze Weaknesses in Accuracy and Speed
if 'accuracy' in df_submission_2.columns and 'speed' in df_submission_2.columns:
    df_submission_2['accuracy'] = df_submission_2['accuracy'].fillna('0').astype(str)
    df_submission_2['accuracy_percent'] = (
        df_submission_2['accuracy'].str.replace('%', '', regex=False).astype(float)
    )
    avg_accuracy = df_submission_2['accuracy_percent'].mean()
    #avg_speed = df_submission_2['speed'].fillna(0).mean()
    df_submission_2['speed'] = pd.to_numeric(df_submission_2['speed'], errors='coerce')
    avg_speed = df_submission_2['speed'].fillna(0).mean()
    insights["Average Accuracy (%)"] = round(avg_accuracy, 2)
    insights["Average Speed (seconds)"] = round(avg_speed, 2)
    print(f"Average Accuracy: {avg_accuracy:.2f}%")
    print(f"Average Speed: {avg_speed:.2f} seconds")
else:
    print("Columns 'accuracy' or 'speed' not found in the DataFrame.")

# 2. Diversify Quiz Topics
if 'quiz_id' in df_submission_2.columns:
    unique_quizzes = df_submission_2['quiz_id'].nunique()
    insights["Total Unique Quizzes"] = unique_quizzes
    print(f"Total Unique Quizzes: {unique_quizzes}")
else:
    print("Column 'quiz_id' not found in the DataFrame.")

# 3. Time-Based Challenges
if 'time_taken' in df_submission_2.columns:
    df_submission_2['time_taken'] = df_submission_2['time_taken'].fillna(0)
    avg_time_taken = df_submission_2['time_taken'].mean()
    insights["Average Time Taken (seconds)"] = round(avg_time_taken, 2)
    print(f"Average Time Taken: {avg_time_taken:.2f} seconds")
else:
    print("Column 'time_taken' not found in the DataFrame.")

# 4. Adaptive Learning
if 'score' in df_submission_2.columns:
    df_submission_2['difficulty'] = np.where(
        df_submission_2['score'] > 80, 'Advanced',
        np.where(df_submission_2['score'] > 50, 'Intermediate', 'Beginner')
    )
    difficulty_distribution = df_submission_2['difficulty'].value_counts()
    insights["Difficulty Distribution"] = difficulty_distribution.to_dict()
    print("Difficulty Distribution:")
    print(difficulty_distribution)
else:
    print("Column 'score' not found in the DataFrame.")

# 5. Gamification Elements
if 'user_id' in df_submission_2.columns and 'quiz_id' in df_submission_2.columns:
    leaderboard = (
        df_submission_2.groupby('user_id')['score']
        .sum()
        .sort_values(ascending=False)
    )
    insights["Leaderboard"] = leaderboard.head(5).to_dict()
    print("Top 5 Users by Total Score:")
    print(leaderboard.head(5))
else:
    print("Columns 'user_id' or 'quiz_id' not found in the DataFrame.")

# Saving Insights to CSV File
csv_data = {
    "Metric": [
        "Average Accuracy (%)",
        "Average Speed (seconds)",
        "Average Time Taken (seconds)",
        "Total Unique Quizzes",
    ],
    "Value": [
        insights.get("Average Accuracy (%)", "N/A"),
        insights.get("Average Speed (seconds)", "N/A"),
        insights.get("Average Time Taken (seconds)", "N/A"),
        insights.get("Total Unique Quizzes", "N/A"),
    ],
}
df_insights = pd.DataFrame(csv_data)
df_insights.to_csv("quiz_insights.csv", index=False)
print("Quiz insights saved to quiz_insights.csv")
```

```python
# Gamification: Create Badges for Achievements
def assign_badges(df):
    # Assign badges based on score thresholds
    df['badge'] = np.where(df['score'] >= 90, 'Gold',
                  np.where(df['score'] >= 75, 'Silver', 'Bronze'))
    return df

df_submission_2 = assign_badges(df_submission_2)
print("User Badges Assigned:")
print(df_submission_2[['user_id', 'score', 'badge']])

# Personalized Recommendations: Suggest Quizzes Based on Performance
def recommend_quizzes(user_id, user_score, all_quizzes):
    # Recommend quizzes of higher difficulty if score is high
    if user_score > 80:
        recommendations = [quiz for quiz in all_quizzes if quiz not in df_submission_2['quiz_id'].unique()]
    else:
        recommendations = all_quizzes[:2]
    return recommendations

all_quizzes = df_submission_2['quiz_id'].unique()
recommendations = {}
for user_id, score in zip(df_submission_2['user_id'], df_submission_2['score']):
    recommendations[user_id] = recommend_quizzes(user_id, score, all_quizzes)

print("Personalized Quiz Recommendations:")
print(recommendations)

# Streaks and Rewards: Track Consecutive Quiz Attempts
df_submission_2['submitted_at'] = pd.to_datetime(df_submission_2['submitted_at'], errors='coerce')
df_submission_2['streak'] = df_submission_2.groupby('user_id')['submitted_at'].diff().dt.days.eq(1).cumsum()

print("User Streaks for Consecutive Participation:")
print(df_submission_2[['user_id', 'submitted_at', 'streak']]) # Also updated here

# Leaderboard: Encourage Competition
leaderboard = df_submission_2.groupby('user_id')['score'].sum().sort_values(ascending=False)
print("Leaderboard (Top Users by Total Score):")
print(leaderboard.head())

# Notifications: Notify Users with Custom Messages
def notify_user(user_id, message):
    print(f"Notification sent to {user_id}: {message}")

# Example notifications based on score and badge
for user_id, badge in zip(df_submission_2['user_id'], df_submission_2['badge']):
    if badge == 'Gold':
        notify_user(user_id, "Congratulations! You've earned a Gold badge! Keep up the great work!")
    elif badge == 'Silver':
        notify_user(user_id, "Well done! You've earned a Silver badge. Keep aiming for Gold!")
    else:
        notify_user(user_id, "Great job! You've earned a Bronze badge. Aim for Silver next time!")

# Saving Engagement Insights to a File
engagement_insights = {
    "User Badges": df_submission_2[['user_id', 'badge']].to_dict(orient='records'),
    "Recommendations": recommendations,
    "Leaderboard": leaderboard.head().to_dict(),
}

for key in engagement_insights:
    if isinstance(engagement_insights[key], dict):
        for subkey in engagement_insights[key]:
            if isinstance(engagement_insights[key][subkey], np.ndarray):
                engagement_insights[key][subkey] = engagement_insights[key][subkey].tolist()
    elif isinstance(engagement_insights[key], np.ndarray):
        engagement_insights[key] = engagement_insights[key].tolist()

# Save insights to a JSON file
with open("engagement_insights.json", "w") as f:
    json.dump(engagement_insights, f, indent=4)

print("Engagement insights saved to engagement_insights.json")
```

```python
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

#Data Preprocessing
# Ensure that 'accuracy', 'time_taken', and 'score' are numeric
df_submission_2['accuracy_percent'] = df_submission_2['accuracy'].str.replace('%', '').astype(float)
df_submission_2['time_taken'] = df_submission_2['time_taken'].fillna(0)

# Feature Selection: relevant features like score, accuracy, speed (time taken), and consistency
features = df_submission_2[['score', 'accuracy_percent', 'time_taken']]

# Clustering: KMeans for clustering students
n_clusters = min(4, features.shape[0])

# Check if the features data has any variance (if any feature has std = 0 then don't cluster)
if features.std().sum() > 0:
    kmeans = KMeans(n_clusters=n_clusters)
    df_submission_2['persona'] = kmeans.fit_predict(features)
else:
    # Handle the case where clustering is not possible (e.g., assign all to one cluster)
    df_submission_2['persona'] = 0

# Persona Labels: personas based on performance metrics
def assign_persona(row):
    if row['score'] > 80 and row['accuracy_percent'] > 90:
        return 'High Performer'
    elif row['score'] < 50:
        return 'Struggling Student'
    elif row['time_taken'] > df_submission_2['time_taken'].mean():
        return 'Procrastinator'
    elif row['accuracy_percent'] < 60:
        return 'Needs Improvement'
    else:
        return 'Steady Performer'

df_submission_2['persona_label'] = df_submission_2.apply(assign_persona, axis=1)

# Strength/Weakness Labeling: based on persona-related conditions
def define_strengths_weaknesses(row):
    if row['score'] > 80:
        return 'Quiz Master' if row['accuracy_percent'] > 90 else 'High Achiever'
    elif row['score'] < 50:
        return 'Underperformer'
    elif row['time_taken'] > df_submission_2['time_taken'].mean():
        return 'Procrastinator'
    elif row['accuracy_percent'] < 60:
        return 'Needs Improvement'
    else:
        return 'Steady Performer'

df_submission_2['strength_weakness_label'] = df_submission_2.apply(define_strengths_weaknesses, axis=1)


print("Persona and Strength/Weakness Labels:")
print(df_submission_2[['user_id', 'persona_label', 'strength_weakness_label']])
```
